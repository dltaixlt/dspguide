# 第一章 数字信号处理的宽度和深度

---
数字信号处理（DSP）是二十一世纪改变科学和工程的技术之一。革命性的改变在大范围内发生：通信，医学影像，雷达和声纳，高保真音乐，石油勘探等等。每个领域都在深度应用DSP技术的算法，数学基础和特定技术。个人很难掌握如此广度和深度的DSP技术。DSP教育包含两方面，不同应用领域的共同基本概念，各个领域的专业技术。本章通过一些DSP在不同领域的精彩应用来引领大家进入DSP的世界。变革已经开始。  
***

### DSP的源头
在计算机科学中，DSP区别于其他技术领域的是其处理对象只有一种：信号。大多数情况来源于真实世界采集的数据，地震波，图像，声音等。DSP就是用数学，算法来处理这些数据的数字化形式。其中包含各种目的，图像增强，语音识别和生成，数据的压缩，存储和传输等。假设我们给电脑添加一个模数转换器，采集了一些真实世界的数据。DSP告诉我们：下一步该做什么。    
DSP的起源从1960年到1970年之间电脑刚发明出来就开始了。那是的电脑是昂贵的，DSP仅有几个特别的应用。先行者的努力集中在4个方面：雷达与声纳，涉及国家安全；石油勘探，能创造大量财富；空间探索，这些数据是不可替代的；医疗影像，能拯救生命。1980年到1990年期间个人电脑革命开始，使得DSP应用有了新的扩展。在军队和政府需求之外，DSP突然被商业市场所驱动。任何人想要从快速扩张领域中赚钱，都开始兜售DSP。DSP通过不同产品来到大众面前：移动电话，CD播放器和电子语音信箱。  
技术的变革是自上而下的。1980年代初，DSP是电子工程系研究生课程，十年之后，DSP成为大学本科课程。今天，DSP是许多领域的科学家和工程师的必备技能。DSP的发展可与前一次电子技术革命相类似。在电子工程领域，每个科学家和工程师都要有电路设计的背景知识，否则会在技术世界中迷失。DSP在未来也是同样。  
当你遇到一个DSP相关问题，去教材或其他出版物中寻找答案时，你通常会遇到一页又一页的公式，晦涩的数学符号和陌生的术语。真是噩梦。很多有DSP经验的人对相关文献都会感到困扰。这并不是说材料本身有问题，只是它面对的受众不同。那些研究者需要通过数学细节来了解其理论含义。  
本书的基本前提是许多DSP实践技术并不需要受到详细的数学和理论的阻碍就能学会。本书是写给那些将DSP用作一个工具，而非将DSP作为一个新职业的人。  
本章剩下部分将介绍几个DSP带来革命性改变的领域。在每个应用的介绍中，你会发现DSP是跨学科的，很依赖于相邻学科领域的技术工作。DSP与以下学科的边界都是模糊和重叠的，通信理论/数值分析/概率与统计/模拟信号处理/判决理论/数字电子技术/模拟电子技术。  

### 电信
电信是将信息从一个地方传输到另一个地方。包含多种信息：电话/电视/文件等。传输首先需要两地之间一个信道（*channel*）。这个信道可以是电线/无线电/光纤等。电信公司收费为用户传输信息，花钱建设和维护信道。其中商业原则很简单：单个信道上传输信号越多，付费越高。DSP革新了电信的多个领域：信号音（signaling tone）的产生和检测，频带迁移，通过滤波去除电流杂音等。这里给出电话网络的三个特别例子：复用（multiplexing），压缩和回声控制。

#### 复用
世界上大概有十亿个电话，你只需要按几个键，交换网络就可以在几秒钟内连接其中任意两个。完成这么巨大的任务是不可思议的！直到1960年代，两个电话的语音连接还需要手工切换和放大。一条连接需要一对线路。而DSP技术将语音信号转换为数字信号流。比特很容易交错（intertwined）传输然后再分离，许多路电话可以在同一条信道上传输。例如，T-carrier系统可以同时传输24条语音。一条语音信号按每秒8000次采样，每个采样值用8bit表示（对数压缩），转换为数字信号。这样一条语音信号速率为64000bits/s，24条语音共1.544Mbits/s。数字传输在商业上的优势是巨大的。模拟交换是昂贵的，数字交换机便宜的多。

#### 压缩
每秒8000点数字采样的语音信号中包含大量冗余，每个样点携带的信息与相邻样点大部分重复。许多DSP算法设计来用更小速率将数字信号转化为码流。这些称为数据压缩算法。配合解压缩算法可恢复原来的信号。这些算法在压缩性能和语音质量上有区别。一般来说，将64kbits/s信号压缩到32kbits/s，语音质量没有太大损失；而压缩到8kbits/s就能明显感受到差别，但是仍然可以用于长距离电话网络。甚至高达2kbit/s的压缩还能用于军事和海底通信，尽管这时语音已经严重失真。

#### 回声控制
长距离电话网络中回声是一个严重问题。通话过程中，你的语音信号传输到远端，其中一部分回传成为回声。距离短于几百英里时，回声时延只有几毫秒，人耳可以习惯。当距离增大，回声变得明显和不堪忍受。在国际通话中，时延可长达几百毫秒，这时相当令人讨厌的。DSP算法测量回声信号，然后生成反向信号用于抵消回声。同样的技术可以用于电话外放时阻止声音回馈产生啸叫。也可以用于数字降噪降低环境噪声。

### 音频处理
人类两大主要感知是视觉和听觉。对应来说，DSP主要应用于图像处理和音频处理。人的听觉内容主要是语音和音乐。DSP在这两方面都有革命性的改变。

#### 音乐
从音乐制作者的麦克风到听众的扬声器的路径相当长。相对于传统模拟信号存储和制作，用数字信号可以减小信号损失。类似于大家比较熟悉的磁带与CD的质量差异。一般来说，录音室中采用多轨录制音频片段，其中包括，独立的人声，乐器录制。这给了音频工程师更灵活的制作空间。将这些独立音轨合成最终作品的过程称为*混音*。DSP给混音提供了许多重要的功能：滤波/信号加减/信号编辑等。  
一个有趣的DSP功能是*人工混响*。如果将独立声轨简单混合，声音听起来会单薄平淡，就像在旷野演奏。人们习惯于环境的混响，而混响是每个录音室要尽量减小的。DSP可以提供人工混响，在混音过程中加入不同的混响来模拟理想的听音环境。几百毫秒的混响可以模拟大教堂的录音，10～20毫秒的混响可以模拟中等大小的听音室。  

#### 语音合成
语音合成和语音识别常用于人机交互。当你的手和眼睛被其他事情占据时，需要用嘴和耳朵替代手和眼睛。比如：开车时，做手术或战场上开枪时。电脑有两种方式合成语音，*数字录音*和*声道模拟*。数字录音是将语音转化为数字方式，压缩存储。回放时再解压转化为模拟信号。一个小时的语音可以用大概3兆大小文件存储，低配置电脑也能胜任。这时常用的语音合成方式。  
声道模拟要复杂的多，它尝试模拟人的发声方式。人的声道是一个发声体，其大小和形状决定共振频率。语音的产生有两种途径：*浊音（Voiced）*和*摩擦音（fricative）*。浊音是声带振动带动声道内空气产生类似周期性的冲激信号。而摩擦音是狭小空间内空气颤动产生的噪声，比如来源于牙齿/嘴唇的开闭。声道模拟应用这两种类型激励信号。声道中的共振和反射通过类似效果的数字滤波器来模拟。  

#### 语音识别
相对于语音合成来说，语音识别更加困难。语音识别是个典型的例子表明人可以完成的很好，而电脑很难完成。电脑可以存储和读取大量数据，一眨眼的功夫完成数学运算，大量重复任务而不厌倦和降低效率，但是面对真实感知数据仍然很难处理。    
DSP将语音识别问题分为两步：特征提取和特征匹配。首先语音信号分割为单字，单字再分析出激励信号和谐波频率。这些特征同预分析的字词特征相比较，寻找最匹配的字词。

### 回声定位
获取远处物体信息的通常方法是测量其反射波。例如，雷达发射脉冲无线电波，接收和测量飞行器的反射波；声纳在水中发射声波来探测潜艇；地质科学家制造地下爆炸，测量反射波来探测深埋地底的岩层信息。尽管这些应用有着共同的线索，但各自都有特别的问题和需求。DSP在这三方面都有革命性的改变。

#### 雷达
Radar （RAdio Detection And Ranging）。雷达系统可以简化描述为，一个无线电波发射器，生成几毫秒的脉冲信号，通过定向天线以光速向外发射。信号遇到飞机后，少部分被反射回来，接收天线获取反射信号。通过发射和接收时间差来计算飞机距离，通过接收天线指向来判定飞机方位。  
雷达系统的控制范围取决于两个因素：脉冲信号的能量和接收天线的噪声大小。不幸的是，要想增加脉冲信号能量就要增加脉冲信号时间长度，这就使得测量接收信号的准确性和精确度变差。因此，雷达系统扫描范围和精确测量之间存在矛盾。  
DSP在上述三个应用中有革命性改变都基于测量范围和精度之间的基本矛盾。首先，DSP可以压缩接收到的脉冲信号，在不损失测量范围情况下提供更准确的测量精度；第二，DSP可以通过滤波器减小接收信号噪声，这同样可以在不损失测量范围情况下提升精度；第三，DSP可以快速生成不同形状和长度的脉冲信号，以适配特殊探测问题。这些算法都需要在高达几百兆赫兹采样率下进行，需要雷达系统设计高速硬件来运行算法。  

#### 声纳
Sonar（SOund NAvigation and Ranging）声纳系统分为两类：主动/被动。主动声纳发出2kHz～40kHz声波，接收和分析回声。应用范围：海底物体探测和定位，导航，通信和海床地图绘制。最大探测范围10～100公里。被动声纳只接收水底的各种声音，包括：自然湍流，海底生物，潜艇和海面船只等的声音。被动声纳不释放能量，很适合隐蔽自己。最重要的应用是军事监测，探测和跟踪潜艇。被动声纳一般监测频率低于主动声纳，频率较低声波的能量不易被水吸收。监测范围可达数千公里。  
同雷达类似，DSP应用在脉冲生成，脉冲压缩和探测信号滤波等方面。一方面，声纳比雷达要简单，因为其处理频率较低。另一方面，声纳又比雷达要难，因为海底环境不一致不稳定。声纳系统通常采用阵列发射和接收信号，经适当控制和混合多通道信号，声纳系统可以定向发送和检测信号。要处理多通道信号，声纳就需要和雷达相同的计算能力

#### 反射地震学
上世纪20年代早期，地质学家用声波探测地壳结构。勘探者制造爆炸，然后记录从地底10公里以内的回声。人们观察这些记录，分析绘制地壳结构。反射地震方法迅速成为石油和矿藏勘探的基本方法，保留至今。  
理性情况下，每层地壳只返回一个回声。但实际情况没有这么简单。每个回声返回时通过每一层地壳都会在地壳间反弹，形成回声的回声，使得传回地表的回声变得复杂，难以解释。DSP在上世纪60年代被广泛使用于分离出基本回声。那么没有DSP的时候人们如何处理？答案是简单的，人们只寻找地质结构简单的地方勘探，减少回声复杂度。DSP可以在复杂的地质环境下寻找石油，比如海底。
